```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['MEDV'])
y = df['MEDV']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)

X_train.shape, X_test.shape, X.shape
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

num_pipeline = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler()
)

X_train_prepared = pd.DataFrame(num_pipeline.fit_transform(X_train), columns=X_train.columns)
X_test_prepared = pd.DataFrame(num_pipeline.transform(X_test), columns=X_test.columns)

X_train_prepared.head()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_pipeline = make_pipeline(
    SimpleImputer(strategy='median'),
    PolynomialFeatures(degree=2, include_bias=False),
    StandardScaler()
)

X_train_poly = pd.DataFrame(poly_pipeline.fit_transform(X_train), columns=poly_pipeline.get_feature_names_out())
X_test_poly = pd.DataFrame(poly_pipeline.transform(X_test), columns=poly_pipeline.get_feature_names_out())

X_train_poly.shape, X_test_poly.shape
```

```python
X_train_poly.head()
```

```python
from sklearn.linear_model import LinearRegression

from sklearn.pipeline import Pipeline

lin_reg_pipeline = Pipeline(
  [
      ('scaler', StandardScaler()),
      ('lin_reg', LinearRegression())
  ]
)

lin_reg_pipeline.fit(X_train, y_train)
print(f'Linear Regression R2 score: {lin_reg_pipeline.score(X_test, y_test)}')
```

```python
poly_pipeline = Pipeline(
    [
        ('poly', PolynomialFeatures(degree=2, include_bias=False)),
        ('scaler', StandardScaler()),
        ('lin_reg', LinearRegression())
    ]
)

poly_pipeline.fit(X_train, y_train)

print(f'Polynomial Regression R2 Score: {poly_pipeline.score(X_test, y_test)}')
```

```python
poly_pipeline.named_steps["lin_reg"].coef_
```

```python
poly_pipeline.named_steps["lin_reg"].coef_ # k parameters = 104
poly_pipeline.named_steps["lin_reg"].intercept_ # b parameters = 1
# total number of parameters (w) = 104 + 1 = 105
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression

degree_list, score_list = range(2, 8), []

for d in degree_list:
  pipe = Pipeline(
      [
          ('poly', PolynomialFeatures(degree=d, include_bias=False)),
          ('scaler', StandardScaler()),
          ('lin_reg', LinearRegression())
      ]
  )
  pipe.fit(X_train, y_train)
  score_list.append(pipe.score(X_test, y_test))

df_results = pd.DataFrame({'degree': degree_list, 'score': score_list}).sort_values(by='score', ascending=False).head(10)


plt.plot(degree_list, score_list)
plt.plot(df_results.iloc[0]['degree'], df_results.iloc[0]['score'], 'ro')
plt.axvline(x=df_results.iloc[0]['degree'], color='red', linestyle='--')
plt.show()

print('\n')
df_results.head(10)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score

ridge_pipe = Pipeline(
    [
        ('poly', PolynomialFeatures(degree=2, include_bias=False)),
        ('scaler', StandardScaler()),
        ('ridge_reg', Ridge(alpha=1.0))
    ]
)

ridge_pipe.fit(X_train, y_train)
print('Ridge score with poly on test data: ',ridge_pipe.score(X_test, y_test))
print('cross_val_score of Ridge with poly: ', cross_val_score(ridge_pipe, X_train, y_train, cv=5, scoring='r2'))
```

```python
ridge_reg = ridge_pipe.named_steps["ridge_reg"]
ridge_reg.coef_
```

```python
import numpy as np

alphas, scores = np.concatenate((np.linspace(0.001, 1, num=10, endpoint=False), np.arange(1, 501, 10))), []

for alpha in alphas:
  poly_ridge_pipe = Pipeline(
      [
          ('poly', PolynomialFeatures(degree=2, include_bias=False)),
          ('scaler', StandardScaler()),
          ('ridge_reg', Ridge(alpha=alpha))
      ]
  )
  poly_ridge_pipe.fit(X_train, y_train)
  scores.append(poly_ridge_pipe.score(X_test, y_test))

df_score = pd.DataFrame({'alpha': alphas, 'score': scores}).sort_values(by='score', ascending=False)

plt.plot(alphas, scores)
plt.plot(df_score.iloc[0]['alpha'], df_score.iloc[0]['score'], 'ro')
plt.axvline(x=df_score.iloc[0]['alpha'], color='red', linestyle='--')
plt.show()

df_score.head(10)
```

```python
from sklearn.model_selection import GridSearchCV

poly_scaler_ridge_pipe = Pipeline(
    [
        ('poly', PolynomialFeatures(include_bias=False)),
        ('scaler', StandardScaler()),
        ('ridge', Ridge())
    ]
)

param_space = {
    'poly__degree': np.arange(2, 6),
    'ridge__alpha': [0.01, 0.1, 1, 10, 100, 150]
}

gscv = GridSearchCV(poly_scaler_ridge_pipe, param_grid=param_space, cv=5, scoring='r2')
gscv.fit(X, y)
gscv.best_params_
```
