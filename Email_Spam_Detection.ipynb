```python
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

cv_mnb = Pipeline(
    [
        ('CountVectorizer', CountVectorizer()),
        ('MultinomialNB', MultinomialNB())
    ]
)


tfidfv_mnb = Pipeline(
    [
        ('TfidfVectorizer', TfidfVectorizer()),
        ('MultinomialNB', MultinomialNB())
    ]
)


# X_train = X_train.values.reshape(-1, )
# X_test = X_test.values.reshape(-1, )
# X = X.values.reshape(-1, )

cv_mnb.fit(X_train, y_train)
print('Testing CountVectorizer:', '\n')
print('Score on test data: ', cv_mnb.score(X_test, y_test))

print('\n'*3)


tfidfv_mnb.fit(X_train, y_train)
print('Testing TfidfVectorizer:', '\n')
print('Score on test data: ', tfidfv_mnb.score(X_test, y_test))


print('\n'*4)


print('Countvectorizer:', '\n'*2)
print('MultinomialNB cross_val_scores with CountVectorizer: ', cross_validate(cv_mnb, X, y, cv=5, n_jobs=-1, scoring=['f1', 'roc_auc']))

print('\n'*5)

print('TfidfVectorizer:', '\n'*2)
print('MultinomialNB cross_val_scores with TfidfVectorizer: ', cross_validate(tfidfv_mnb, X, y, scoring=['f1', 'roc_auc'], cv=5, n_jobs=-1))
```
