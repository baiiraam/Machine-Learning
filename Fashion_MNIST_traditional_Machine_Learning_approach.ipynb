# Fashion MNIST traditional Machine Learning approach

```python
import tensorflow.keras.datasets.fashion_mnist as fashion_mnist

fashion = fashion_mnist.load_data()
```

```python
(X_train, y_train), (X_test, y_test) = fashion
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
import matplotlib.pyplot as plt

plt.imshow(X_train[0], cmap='binary')
plt.axis('off')
plt.show()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
n_rows = 3
n_cols = 10

plt.figure(figsize=(n_cols*1.2, n_rows*1.2))
for row in range(n_rows):
  for col in range(n_cols):
    index = row * n_cols + col
    plt.subplot(n_rows, n_cols, index + 1)
    plt.imshow(X_train[index], cmap='binary', interpolation='nearest')
    plt.axis('off')
    plt.title(class_names[y_train[index]])

plt.subplots_adjust(wspace=0.2, hspace=0.5)
plt.show()
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)
```

```python
import numpy as np
```

```python
np.unique(y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from lightgbm import LGBMClassifier

models = {'Decision Tree': DecisionTreeClassifier(), 'LGBMClassifier': LGBMClassifier(), 'Random Forest': RandomForestClassifier()}

for name, model in models.items():
  print(f'{name}: ', '\n')

  model.fit(X_train, y_train)
  y_preds = model.predict(X_test)

  labels = [i for i in range(10)]

  f1_scores = f1_score(y_test, y_preds, average=None, labels=labels)
  precision_scores = precision_score(y_test, y_preds, average=None, labels=labels)
  recall_scores = recall_score(y_test, y_preds, average=None, labels=labels)

  f1_ = {label: score for label, score in zip(labels, f1_scores)}
  precc = {label: score for label, score in zip(labels, precision_scores)}
  recc = {label: score for label, score in zip(labels, recall_scores)}

  cm = confusion_matrix(y_test, y_preds, labels=model.classes_)
  disp = ConfusionMatrixDisplay(cm, display_labels=model.classes_)

  keys, values = [], []

  print(f'f1 score for each label on test data: {f1_}', '\n')
  for key, value in f1_.items():
    keys.append(key)
    values.append(value)

  keys, values = np.array(keys), np.array(values)
  print(f'{name} works best for "{class_names[keys[np.argmax(values)]]}" classification. Result: {np.max(values)}', '\n')
  print(f'{name} works poorly for "{class_names[keys[np.argmin(values)]]}" classification. Result: {np.min(values)}', '\n')



  print(f'precision score for each label on test data: {precc}', '\n')
  print(f'recall score for each label on test data: {recc}', '\n')

  plt.figure(figsize=(10, 10))
  disp.plot()
  plt.show()

  print('\n'*3)
```
